```python
from faker import Faker
import pandas as pd

# Initialize a Faker object
fake = Faker()

# Function to generate a single fake profile
def generate_fake_profile():
    profile = {
        "CustomerID": fake.uuid4(),
        "FirstName": fake.first_name(),
        "LastName": fake.last_name(),
        "Email": fake.email(),
        "Phone": fake.phone_number(),
        "Address": fake.street_address(),
        "City": fake.city(),
        "Country": fake.country(),
        "JobTitle": fake.job(),
        "CompanyName": fake.company(),
        "BirthDate": fake.date_of_birth().strftime("%Y-%m-%d")
    }
    return profile

# Function to generate a list of fake profiles
def generate_fake_profiles(count):
    profiles = [generate_fake_profile() for _ in range(count)]
    return pd.DataFrame(profiles)

# Generate 100 fake profiles
num_profiles = 100
fake_profiles_df = generate_fake_profiles(num_profiles)

# Print each profile
print(fake_profiles_df)

# Optionally, save the data to a CSV file
fake_profiles_df.to_csv('fake_profiles.csv', index=False)
```
